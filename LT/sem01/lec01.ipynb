Итак, место питона:

У нас есть два полюса: процессор, родной язык которого - микрокод (ограниченное количество вшитых в конкретную модель процессора команд), и человек, родной язык которого - choose your fighter. Программирование - во многом процесс перевода приказов с человеческого языка на процессорный. 
Уровни кода:
- микрокод (на нем понимает процессор, с ним работают только инженеры и создатели процессоров)
- машинный код (команды состоят из нулей и единиц, на нем уже могут писать программы люди)
- ассемблер (язык низкого уровня: буквенные мнемоники для команд машинного кода)

...
- языки высокого уровня. 

Многоточие потому, что это не полная иерархия, но нам достаточно этого. 

Первоначальный компилятор языка С был написан в ассемблере, а интерпретатор питона (его стандартная реализация, по крайней мере) написан на С. Это означает, что:

а) питон очень дружелюбный, максимально близок к человеческому языку, находится на нашей шкале вплотную к человеку.

б) когда интерпретатор питона переводит команды в машинный язык, их приходится переводить на самом деле несколько раз. Как лингвисты, вы должны знать, насколько несовершенен машинный перевод (сейчас улучшили, но все еще неидеально). Тут то же самое: из-за того, что у нас получается испорченный телефон, программы в питоне работают медленнее, чем программы в С и в ассемблере. 

    (но хороший программист в питоне может написать более быстрый код, чем плохой программист в С)

Вот уже встретились два слова: компилятор и интерпретатор, что это такое и в чем разница? На занятии об этом говорилось, но попробую переформулировать более ясно:

- Если мы говорим, что какой-то ЯП имеет свой компилятор, это означает, что сперва нужно полностью написать программу на этом языке, а потом уже компилятор (переводчик на машинный язык) ее всю сразу скомпилирует - или не скомпилирует, а вывалит кучу непонятных ошибок, и ищи-свищи, что их вызвало. 
- Если какой-то ЯП имеет интерпретатор, это означает, что мы можем запустить его интерпретатор и писать команды одну за другой, а он будет сразу же кидаться их выполнять. 

Одно важное следствие из этого, о котором на занятии было забыто сказать - компилятор переводит вашу программу в такой вид, что ее можно будет запустить без ЯП, на котором она была записана, на любом компьютере с подходящей архитектурой. Если же у вас интерпретатор, это означает, что для запуска вашей программы на другом компьютере на него обязательно понадобится установить этот же интерпретатор (то есть, если вы пишете скрипт на питоне и хотите его показать другу на его компьютере, придется другу устанавливать питон). 

Главные качества питона, за которые его мы с вами изучаем:

- у него очень простой синтаксис (то есть, правила, которым нужно следовать, когда пишешь на нем код)
- он интерактивный, а значит, на нем удобно экспериментировать
- для него написано огромное количество готовых библиотек, которые можно использовать
- большое комьюнити и мириады туториалов.

Что такое среда разработки?

IDE (integrated development environment) - среда разработки - это инструмент, который совмещает в себе текстовый редактор и интерпретатор питона. В текстовом редакторе мы пишем скрипты, а потом IDE позволяет их выполнять. Скрипты можно на самом деле писать хоть в блокноте! Единственное условие - сохранять их в формате .py. (.ipynb - формат тетрадки IPython). Можно жить без IDE, но грустно. Написать скрипт в блокноте мы пробовали на занятии: достаточно написать все, что нужно, в обычном текстовом файле, а при сохранении выбрать тип файла "все типы" или просто потом переименовать так, чтобы вместо .txt было .py. Тогда такой файл можно будет запустить, если перейти в папку с ним в командной строке и набрать в командной строке python hello.py, где hello.py - это название вашего файла. 

В любой среде разработки питона на самом деле есть два режима: интерактивный и сценарный. Что это за режимы?

Интерактивный - это как раз то, что по умолчанию открывается в IDLE, то есть, у вас есть строка с приглашением ввода (>>>), куда вы можете вводить свои команды, и каждый раз, когда вы ввели что-то и нажали Enter, это что-то тут же выполняется. 

Сценарный - это когда вы заводите какой-нибудь файлик и записываете в него все команды по порядку, а потом сохраняете его и запускаете весь целиком.

Во-первых, некоторые особенности работы с тетрадками юпитера: тетрадка состоит из ячеек, которые бывают двух (или трех) видов: Code, Markdown, Raw (последний не везде есть). В ячейках типа Code вы можете писать команды на питоне и запускать их в любой последовательности. Обратите внимание, что ячейки взаимосвязаны: если в одной ячейке вы завели переменную а и запустили ее, то эта переменная будет понятна и в другой ячейке. Все, что вы запускаете, складывается в память и хранится, пока вы не остановите кернел (кернел - "ядро", тот самый интерпретатор питона, который во время вашей работы с тетрадкой запущен и ловит ваши команды). 

В ячейках типа Markdown можно писать форматированный текст, так, как я это делаю теперь, текст можно делать **жирным**, *курсивом*, можно писать формулы и всякое прочее. В интернете существует довольно большое количество туториалов по маркдауну юпитера. Осваивать прямо все я не призываю (мы все-таки не версткой и дизайном занимаемся), но для красоты имейте в виду. :)

Также в тетрадках есть стандартные горячие клавиши, которыми удобно пользоваться. 

Если у вас не мигает курсор, а просто выделена ячейка (выделенная ячейка обретает синюю полоску слева):

B - для создания новой ячейки

C - копирует выделенную ячейку

V - вставляет скопированную ячейку

Y - меняет тип ячейки на код

M - меняет тип ячейки на маркдаун

D (нужно нажать два раза подряд) - удаляет ячейку

Shift+Enter - запускает ячейку с кодом (или рендерит маркдаун). 

Пользоваться горячими клавишами необязательно, но обычно очень удобно.

В питоне все - это объекты. Немного по-другому нужно работать со строками. Если вы хотите сохранить в свою переменную строчку текста, нужно ее взять в кавычки. Кавычки любого вида (главное парные) сообщают питону, что у него тут строка. 

```python
string = 'строка'
print(string)
```

Питон знает многообразие кавычек (у них есть небольшие различия, о них еще поговорим):

'string'

"string"

'''string'''

Также мы с вами узнали две *функции*: для ввода и вывода информации. Они выглядят и работают примерно так:

Кстати, обратите внимание, что если у вас несколько команд подряд, вы просто пишете их каждую на новой строке и все. Никаких знаков между ними (например, ; ) ставить не нужно.
```python
num = input('Введите число: ')
num
```

Обратите внимание: хотя я ввела число, функция input() все равно считала его как строку. Об этом тоже еще будем подробнее говорить. Чтобы дать питону знать, что мы ожидаем число, нужно дополнительно написать:

```python
num = int(input())
```

Про это я вам подробно рассказать уже не успела, но если разбирать эту конструкцию, то она устроена примерно следующим (луковичным) образом:

1. Сперва питон обнаруживает функцию input(). Не забывайте про скобочки! Хотя они вроде бы пустые, они сообщают питону, что мы имеем в виду *функцию*, а не просто ее название. Что будет, если вы забудете скобочки:

```python
num = input
num
```

Ух, жуть какая. В общем, не забывайте. Потом объясню, что это такое, когда до функций дойдем. 

2. Обнаружив функцию, питон ее тут же выполняет (потому что она внутри всей луковки). Функция считывает от пользователя какие-то символы на вводе и передает их дальше в функцию int() - да-да, это тоже функция! Она превращает все, что ей запихнуть в скобочки, в числа. В смысле, в числа в мировоззрении питона, такие, которые он умеет складывать, вычитать и так далее. 

3. Ну и наконец, когда функция int(), приняв ваше введенное чиселко в виде строки, вернула его же в виде числа, знак = сообщает питону, что нужно положить это чиселко по адресу num. Все, готово, в адресной книжке запущенного питона заведена новая запись.
