## Встроенные типы данных. Числовые типы данных. Арифметические операторы. Условные операторы. If-Else Statement.

Итак, питон работает с объектами. Все, что есть в этом мире - объекты питона! Объекты (реальные) хранятся где-то в оперативной памяти компьютера, а питон при запуске скрипта или интерактивного режима заводит свою телефонную книжечку, куда записывает адреса всех этих объектов. Допустим, у нас есть строчка:

    x = 5 

Здесь 5 - реальный конкретный объект (число), который компьютер складывает в оперативную память (как он это делает, нам даже думать не нужно! В отличие от бедненьких программистов С). Знак "=" просит питон сохранить адрес этого объекта у себя, а для нашего удобства этот адрес записывается как переменная х. По аналогии с телефонной книжкой: ваша подружка может быть в соседнем городе, но ваш телефон хранит ее номер, который вы, в свою очередь, не помните, но записали его под ее именем. С переменными и объектами в питоне точно так же!

Знак "=" еще называется оператор присваивания. 

Объекты, как уже было сказано, делятся на разные типы. Как в жизни: у нас есть люди, стулья, котики и так далее. Компьютеру важно знать, с каким типом объектов он работает, потому что у него внутри все представляется цифрами, и для него что буква, что цифра - все равно. Питон определяет тип конкретных объектов самостоятельно, но нам нужно знать, какие бывают встроенные типы, чтобы понимать, какие можно с ними производить действия. 

Понятия **тип** и **класс** в питоне по сути тождественны: это синонимы. В других ЯП это не так!

Когда я говорю "встроенные" - это означает, что эта вещь уже реализована в стандартном питоне. Бывают пользовательские классы: те, которые программисты создают сами, мы научимся это делать. Бывают и пользовательские (сторонние) библиотеки: уже готовые инструменты, которые можно использовать. 

Мы рассмотрели некоторые неизменяемые типы. Неизменяемые означает то, что эти объекты нельзя изменить по кусочкам, только перезаписать полностью. Мы можем написать так:

    x = 5 # мы присвоили переменной х объект "5"
    x += 1 # оператор += одновременно прибавляет к х единицу и записывает результат обратно в х

При этом на самом деле пятерочка останется валяться где-то в памяти, но мы потеряем к ней доступ; у питона есть сборщик мусора: такое устройство, которое собирает все наши потерянные пятерочки и удаляет их за нами из памяти, чтобы потом другие программы могли использовать эту память. Во второй строчке создастся новый объект "6", который теперь и будет храниться по адресу в х. 

Для работы с числами в питоне есть следующие (неизменяемые) типы:
- int (целые числа)
- float (числа с запятой)
- complex (комплексные числа, с мнимой и действительной частью)
- bool (подвид int: логическое значение False/True)

Для работы со строками есть тип str.

Для того, чтобы явно преобразовать объект из одного типа в другой, можно использовать одноименные функции, например:

```python
x = 5
print(type(x))
float(x)
```

Неплохо помнить, что при работе с нецелыми числами могут возникать разного рода непонятности, например:

```python
0.008 + 0.1
```

Это происходит потому, что мы считаем в десятичной системе, а компьютеру приходится все наши десятичные дроби переводить в двоичный вид. В какой-то момент там и возникает такая же история, как с дробью 1/3, которая равна 0.333333....., и компьютер округляет получившуюся фигню, потому что она у него не помещается.

### Арифметические операторы

В принципе все математические штуки воплотимы в питоне: точно так же умножение и деление выполняются в первую очередь, а сложение и вычитание во вторую, и так далее. У нас есть стандартный набор:

    +

    -

    *

    / - обязательно неявно преобразует целые числа к float! вернет дробь, даже если делится нацело. 

    // - целочисленное деление

    % - остаток от деления

    ** - степень

Скобки работают как и в школьной математике!

К любому из этих операторов можно приписать =, например, +=: тогда оператор не будет возвращать новое значение, а запишет в старую переменную (как выше в примере про x += 1). Обычно операторы += и -= используются для счетчиков. 

### Логические операторы

    > - больше

    < - меньше

    >= - больше или равно

    <= - меньше или равно

    == - равно

    != - не равно

    and - логическое "и"

    or - логическое "или"

    not - логическое отрицание
    
    in - проверка членства (символ 'a' в строке 'hello' => 'a' in 'hello)

Скобки опять же работают!

Возьмите на заметку: выражения с логическими операторами питон вычисляет и превращает в объекты типа bool!

А также не забывайте, что любой практически объект можно превратить в bool. Если объект - число, то все, кроме 0, превратится в True, а ноль - в False. Если объект - строка, то пустая строка превратится в False, непустая - в True. 

Не путайте "=" с "==". 

```python
bool('False') # как превратить ложь в истину? 
```

### Условные инструкции



    if <condition>:
        command
        command
    elif <condition>:
        command
    elif <condition>:
        command
    ...
    else:
        command

Это - общий шаблон для условных инструкций. Elif и else могут отсутствовать, обязательный только if, после которого должна быть хотя бы одна команда. if и elif могут принимать в качестве условия какое-нибудь выражение (a == b) или объект ('abc'); выражение при этом вычисляется и дает в качестве результата bool, а объект неявно преобразовывается в bool по описанным выше правилам. 

Поэтому грамотная и часто используемая проверка:

    if a:
        <do something>

Отступы перед командами и после двоеточия называются индентингом и сообщают питону о том, что эти команды находятся *внутри* какой-то конструкции, в нашем примере это операторы if-elif-else. То, что находится внутри, называется телом. Команды, которые находятся в теле if, выполняются только в том случае, если выполняется сам if. Индентинги - это важное понятие в мире питона, в отличие от других ЯП, здесь они обязательны.
